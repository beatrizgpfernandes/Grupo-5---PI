```python
import pandas as pd
import matplotlib.pyplot as plt


# Coloque o arquivo csv na mesma pasta e digite seu nome. É possível digitar o caminho do csv também
leitor = pd.read_csv('linhas.csv')
```

```python
# 1 -Rode o comando e visualize as colunas que deseja plotar no próximo passo  
print(leitor.head())
print(leitor.columns)
```

```python


plt.figure(figsize=(10,6))

# Em 'plt.scatter' coloque o nome das colunas que deseja plotar (tem q ser exatamente o nome que aparece na hora do print, no passo anterior)
# É possível mudar o tipo de gráfico. Em vez de 'plt.scatter', pode-se usar 'plt.bar', 'plt.plot', 'plt.pie'...
plt.scatter(leitor['CpuID'], leitor['cpuPercent'])


plt.title('Gráfico de métricas')
plt.xlabel('Tempo')
plt.ylabel('Uso componente (%)')
plt.grid(True)

plt.show()
```
